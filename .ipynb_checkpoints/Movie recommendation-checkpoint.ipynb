## Text Embedding

```python
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

words = []

for sent in df['combined']: 
    words.append(simple_preprocess(sent))
```

```python
import gensim
model = gensim.models.Word2Vec(words)
```

```python
#  checking is there is any empty vector or not 
emp_vec = 0 
for doc in words:
    vec = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if not vec:
        emp_vec += 1
print(emp_vec)
```

```python
avg_vec = []
for doc in words:
    vec = [model.wv[word] for word in doc if word in model.wv.index_to_key]
    if not vec:
        avg_vec.append(np.zeros(model.vector_size))
    else:
        avg_vec.append(np.mean(vec,axis=0))
```

```python
x_new = np.array(avg_vec,dtype=float)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(x_new)
```

```python
def recommend(title): 
    title = title.lower().strip()
    ind = df[df['title'].str.lower() == title]
    if len(ind) == 0:
        print('This movies is not present in our dataset so we cant give you recommendation. ')
    else:
        ind = ind.index[0]
        scores = list(enumerate(similarity[ind]))
        top_scores = sorted(scores,key=lambda x:x[1],reverse=True)[1:6]
        for i in top_scores: 
            print(df.iloc[i[0]]['title'])
        
```

```python
recommend('Dangal')
```
